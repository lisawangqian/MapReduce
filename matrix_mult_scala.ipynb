```scala
val n: Int = 6
val g: Int = 3
val gSize: Int = n/g

val matrix1 = sc.textFile("matrix1")
val matrix2 = sc.textFile("matrix2")
```

```scala
def linesToRowColumn(l: String): ((Int, Int), Double) = {
    val ll = l.split(" ")
    return((ll(0).toInt - 1, ll(1).toInt - 1), ll(2).toDouble)
}

def AssignGroups(l:((Int, Int),Double), left:Boolean): 
    List[((Int, Int), ((Int, Int), Int, Double))] = {
    
    val k = l._1
    val v = l._2
    val i = if(left) k._1/gSize else k._2/gSize
    val m = if(left) 1 else 2
    
    def go(group: Int, acc: List[((Int, Int), ((Int, Int), Int, Double))]): 
        List[((Int, Int), ((Int, Int), Int, Double))] = {
        
        if(group >= g) return acc
        else return go(group+1, 
                       acc ++ List((if(left) 
                                        (i, group) 
                                    else 
                                        (group, i), 
                                    (k, m, v))))
    }    
    return go(0, List())
}

val matrix1Map = matrix1.map(linesToRowColumn).flatMap(
    x => AssignGroups(x, true))
val matrix2Map = matrix2.map(linesToRowColumn).flatMap(
    x => AssignGroups(x, false))

matrix1Map.collect().foreach(println)
matrix2Map.collect().foreach(println)
```

```scala
def reduceConvert(l:List[((Int, Int), Int, Double)]): 
    List[(Int, (Int, Double, Int))] = {
    def go(group: Int, acc: List[(Int, (Int, Double, Int))]): 
        List[(Int, (Int, Double, Int))] = {
        
        if(group >= n*2*gSize) return acc
        else return go(group+1, 
                       acc ++ List(if(l(group)._2 == 1)
                                        (l(group)._1._2, (l(group)._1._1, l(group)._3, 1))
                                   else
                                        (l(group)._1._1, (l(group)._1._2, l(group)._3, 2))))
    }    
    
    val mat = go(0, List())
   
    return mat
   
}  

def reduceGroup(l:List[(Int, (Int, Double, Int))]): 
    Map[Int,List[(Int, (Int, Double, Int))]] = {
    return l.groupBy(t => t._1)
    
}
    

val unionMatrix = matrix1Map.union(matrix2Map)
val matrixGroup = unionMatrix.groupByKey().values.map(i => i.toList)
val result_sub1 = matrixGroup.map(x => reduceConvert(x))

var result_sub2 = result_sub1.map(x => reduceGroup(x))
result_sub2.collect().foreach(println)

```

```scala
import scala.collection.mutable.ListBuffer
def reduceSum(m:Map[Int,List[(Int, (Int, Double, Int))]]): 
  List[((Int, Int), Double)]={
    var i = 0 
    var col = new ListBuffer[((Int, Int), Double)]()
    var mapKey = m.keys
    def reduceProduct(l:List[(Int, Double, Int)]): 
        ((Int,Int), Double) = {
        var i = l(0)._1
        var j = l(1)._1
        var pro = l(0)._2 * l(1)._2
        return ((i,j), pro)
    }
    
    for (i <- mapKey){
        var element = m(i).map(t => t._2)
        var element_comb = element.toSeq.combinations(2).filter(t=> t(0)._3 < t(1)._3)
        var ele_pro = element_comb.map(x=>reduceProduct(x.toList))
        ele_pro.foreach(x => col+=x)
    }
    
    val colList = col.toList
    val sum_g = colList.groupBy( _._1).map( kv => (kv._1, kv._2.map( _._2).sum)).toList
    return sum_g
    
                                                
}

var result_sub3 = result_sub2.map(x => reduceSum(x))
var finalResult = result_sub3.flatMap(x=>x)
finalResult.collect().foreach(println) 
```
